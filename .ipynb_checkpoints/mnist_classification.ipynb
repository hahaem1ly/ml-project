```python
import numpy as np                      # For numerical operations
import tensorflow as tf                 # For building and training the model
from tensorflow.keras.models import Sequential   # To create a sequential model
from tensorflow.keras.layers import Input, Flatten, Dense  # Model layers
import matplotlib.pyplot as plt         # For visualizing the data

```

```python
# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Check the initial shapes of the data
print(f"Shape of X_train: {X_train.shape}")  # Expected: (60000, 28, 28)
print(f"Shape of X_test: {X_test.shape}")    # Expected: (10000, 28, 28)
print(f"Shape of y_train: {y_train.shape}")  # Expected: (60000,)
print(f"Shape of y_test: {y_test.shape}")    # Expected: (10000,)
```

```python
# Visualize the first 5 training images with their labels
fig, axes = plt.subplots(1, 5, figsize=(12, 3))
for i, ax in enumerate(axes):
    ax.imshow(X_train[i], cmap="gray")  # Show the image in grayscale
    ax.set_title(f"Label: {y_train[i]}")  # Display the corresponding label
    ax.axis("off")  # Remove axes for cleaner visualization
plt.show()
```

```python
# Normalize pixel values to [0, 1]
X_train = X_train / 255.0
X_test = X_test / 255.0

# Confirm shapes after normalization
print(f"Shape of X_train after normalization: {X_train.shape}")  # (60000, 28, 28)
print(f"Shape of X_test after normalization: {X_test.shape}")    # (10000, 28, 28)
```

```python
# Convert labels to one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# Confirm shapes after one-hot encoding
print(f"Shape of y_train after one-hot encoding: {y_train.shape}")  # (60000, 10)
print(f"Shape of y_test after one-hot encoding: {y_test.shape}")    # (10000, 10)
```

```python
# Build the neural network
model = Sequential([
    Input(shape=(28, 28)),           # Input layer: 28x28 images
    Flatten(),                      # Flatten the 2D images into a 1D vector
    Dense(128, activation='relu'),  # Hidden layer with 128 neurons
    Dense(64, activation='relu'),   # Hidden layer with 64 neurons
    Dense(10, activation='softmax') # Output layer with 10 neurons for 10 classes
])

# Print a summary of the model
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',                   # Optimizer to update weights
              loss='categorical_crossentropy',    # Loss function for classification
              metrics=['accuracy'])               # Metric to track performance
```

```python
# Train the model
history = model.fit(X_train, y_train,             # Training data
                    validation_split=0.1,         # Use 10% of training data for validation
                    epochs=5,                     # Number of iterations over the data
                    batch_size=32)                # Number of samples per batch
```

```python
# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot training and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
